# Fractional Brownian motion simulation

A simple (but inefficient) algorithm: simulate from the exact multivariate normal distribution.

Recall that if $B^H$ is an FBM, then
$$
\mathrm{cov}(B^H_t, B^H_s) = \frac{1}{2} (s^{2H} + t^{2H} - |t-s|^{2H}).
$$

To simulate $B^H_{t_i}$ at points $t_i = i\Delta t$, we simulate from the corresponding normal distribution.


```python
from math import *
import numpy as np
from scipy import stats as st
from matplotlib import pyplot as plt
```

```python
H = 0.3

# grid: t_i = i*t/n, where i = 0,...,n
t = 1.0
n = 4096*4
T = np.linspace(0, t, n+1)
```

```python
# FBM simulation from the multivariate normal distribution

# Covariance matrix
BH_cov = 0.5*(np.add.outer(T**(2*H), T**(2*H)) - np.abs(np.subtract.outer(T,T))**(2*H))

# Random vector (B^H_{t_i}), i=0,...,n
BH = st.multivariate_normal(mean=np.zeros(n+1), cov = BH_cov, allow_singular=True)

# Simulate and plot m paths
m = 1
plt.plot(T, BH.rvs(m).T)
```

# Muravlev's method 

Let $\xi=(\xi_\beta)_{\beta>0}$ be a Gaussian process with zero mean and covariance function
$$
\mathrm{cov}(\xi_\alpha, \xi_\beta) = \frac{1}{\alpha+\beta}, \qquad \alpha,\beta>0.
$$

Let $B_t$ be a Brownian motion independent of $\xi$. Define the family of Ornstein-Uhlenbeck processes $Z^\beta = (Z_t^\beta)_{t\ge 0}$ driven by the same Brownian motion $B_t$:
$$
d Z_t^\beta = -\beta(Z_t^\beta +\xi_\beta) dt + d B_t, \qquad Z_0^\beta = 0.
$$

Then for $H\in(0,\frac12)$ we have the following representation:
$$
B_t^H = c_H \int_0^\infty \beta^{-1/2-H} Z_t^\beta d \beta,
$$
where $c_H$ is some constant.

We simulate this process as follows.
1. Take $\beta_j = j\Delta \beta, j=1,\dots,k$, and simulate the value of $\xi_{\beta_j}$ from the multivariate normal distribution. 

2. Simulate a trajectory of the Brownian motion at points $t_i = i\Delta t$, $i=0,\dots,n$.

3. Simulate $k$ O-U processes with the above $\beta_j$ using Euler's scheme.

4. Approximate the integral by
$$
B_{t_i}^H \approx \sum_{j=1}^k \beta_j^{-1/2-H} Z_{t_i}^{\beta_j} \Delta \beta.
$$

```python
# Muravlev's method
H = 0.1

# Grid of beta_j
dbeta = 0.1
k = 2000
beta = np.linspace(dbeta, dbeta*k, k)
# It may be better to take a non-uniform grid of beta 
# For example:
#beta = np.exp(np.linspace(dbeta, dbeta*k, k))

# Covariance matrix of the process xi
xi_cov = 1/np.add.outer(beta, beta)

# Random vector (xi_{beta_j}), j=1,...,k
Xi = st.multivariate_normal(mean = np.zeros_like(beta), cov = xi_cov, allow_singular=True)

# One trajectory of xi
xi = Xi.rvs(1)
plt.plot(beta, xi)
plt.title("Process xi")
plt.show()

# Increments of the Brownian motion
dt = t/n
dW = np.random.randn(n)*sqrt(dt)

# Trajectories of the OU processes
Z = np.zeros((n+1, beta.size))
for i in range(n):
    Z[i+1] = Z[i] - beta*(Z[i]+xi)*dt + dW[i]

# One trajectory of FBM
BH = np.sum(beta**(-0.5-H) * Z * np.diff(beta, prepend=0), axis=1)
plt.plot(T, BH)
plt.title("Fractional BM")
plt.show()
```

```python
# my Muravlev's method
H = 0.3

# Grid of beta_j
dbeta = 4
k = 4000
beta = np.linspace(dbeta, dbeta*k, k)
# It may be better to take a non-uniform grid of beta 
# For example:
#beta = np.exp(np.linspace(dbeta, dbeta*k, k))

# Covariance matrix of the process xi
xi_cov = 1/np.add.outer(beta, beta)

# Random vector (xi_{beta_j}), j=1,...,k
Xi = st.multivariate_normal(mean = np.zeros_like(beta), cov = xi_cov, allow_singular=True)

# One trajectory of xi
xi = Xi.rvs(1)
plt.plot(beta, xi)
plt.title("Process xi")
plt.show()

# Increments of the Brownian motion
dt = t/n
dW = np.random.randn(n)*sqrt(dt)

# Trajectories of the OU processes
Z = np.zeros((n+1, beta.size))
Z[0] = xi
for i in range(n):
    Z[i+1] = Z[i] - beta*Z[i]*dt + dW[i]

# One trajectory of FBM
BH = np.sum(beta**(-0.5-H) * (Z-xi) * np.diff(beta, prepend=0), axis=1)
plt.plot(T, BH)
plt.title("Fractional BM")
plt.show()
```
