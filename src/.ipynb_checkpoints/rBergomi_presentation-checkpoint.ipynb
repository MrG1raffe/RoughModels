```python
H = 0.25
T = 4
rho = 0
eta = 1
xi = 0.2 ** 2
m = 10**4
n = 1000
S0 = 100
```

```python
%%time
rB = rBergomi(H, T, rho)
Z, W = rB.WZ_sample(m, n)
v = rB.sample_v(W, eta, xi)
S_Bayer = S0 * rB.sample_S(v, Z)
```

```python
plt.plot(np.linspace(0,T,m), S_Bayer)
plt.show()
```

```python
v_last_moment = v[-1,:]
```

## Volterra

```python
%%time
T = 4
n_ = 2000
N = int(np.floor(T * n_) + 1)
eta = 1
xi = 0.2**2
y0 = 0
S0 = 100

mu = 0
sigma = eta * np.sqrt(2*H)

t_grid = np.arange(N) / n_
Y, dW = simulate_stochastic_convolution(T=T, 
                                        n=n_, 
                                        alpha=H - 0.5, 
                                        y0=y0,
                                        mu=mu, 
                                        sigma=sigma, 
                                        kappa=5)
V = xi * np.exp(Y - 0.5 * eta**2 * t_grid**(2*H))
S = price_from_vol(V, mode='var', T=T, x0=S0, rho=-0.4, dW=dW)
```

```python
V_last_moment = []
for i in range(n):
    Y, dW = simulate_stochastic_convolution(T=T, 
                                        n=n_, 
                                        alpha=H - 0.5, 
                                        y0=y0,
                                        mu=mu, 
                                        sigma=sigma, 
                                        kappa=5)
    V = xi * np.exp(Y - 0.5 * eta**2 * t_grid**(2*H))
    S = price_from_vol(V, mode='var', T=T, x0=S0, rho=0, dW=dW)
    V_last_moment.append(V[-1])
    
V_last_moment = np.array(V_last_moment)
```

```python
kstest(v_last_moment, V_last_moment)
```
